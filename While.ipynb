Zgadnij liczbę: komputer losuje liczbę od 1 do 20, użytkownik zgaduje aż trafi, użyj while.

```python
import random
liczba_komp = random.randint(1, 20)
liczba_uzytk = int(input("Zgadnij liczbę od 1 do 20: "))
while liczba_uzytk != liczba_komp:
  print("Zła liczba, spróbuj ponownie!")
  liczba_uzytk = int(input("Zgadnij liczbę od 1 do 20: "))
print("Brawo zgadles")
```

Oblicz sumę wszystkich liczb podanych przez użytkownika aż poda 0 (0 kończy pętlę).

```python
liczby = int(input("Podaj liczby"))
suma = 0
while liczby != 0:
  suma = suma + liczby
  liczby = int(input("Podaj liczby"))

print(suma)
```

Stwórz prostą grę, w której licznik punktów rośnie o 1 co sekundę, ale pętla kończy się, gdy suma punktów przekroczy 100.

```python
import time

punkty = 0

while punkty <= 100:
    punkty += 1
    print("Punkty:", punkty)
    time.sleep(1)  # czekaj 1 sekundę

print("Koniec gry! Twoje punkty przekroczyły 100.")
```

Symulacja bankomatu: użytkownik może wpłacać i wypłacać pieniądze, aż wpisze "wyjdź" – użyj while.

```python
print("Jesli chcesz zakonczyc, wpisz wyjdz /n jesli chcesz wplacic pieniadze wpisz 1 /n jesli chcesz wyplacic pieniadze wpisz 2")
koniec = input("Podaj wartosc")
konto = 0
while True:
    print("\nJeśli chcesz zakończyć, wpisz 'wyjdz'")
    print("Jeśli chcesz wpłacić pieniądze, wpisz 1")
    print("Jeśli chcesz wypłacić pieniądze, wpisz 2")

    koniec = input("Podaj wartość: ")
    if koniec == "1":
        wplata = int(input("Ile chcesz wpłacić? "))
        konto += wplata
        print("Twoje saldo:", konto)

    elif koniec == "2":
        wyplata = int(input("Ile chcesz wypłacić? "))
        if wyplata > konto:
            print("Nie masz tyle pieniędzy na koncie!")
        else:
            konto -= wyplata
            print("Twoje saldo:", konto)

    elif koniec.lower() == "wyjdz":
        print("Koniec programu")
        break

    else:
        print("Niepoprawna opcja, spróbuj ponownie.")




```
