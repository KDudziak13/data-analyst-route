4. Wypisz liczby od 1 do 20, ale dla wielokrotności 3 wypisz "Fizz".

```python
for i in range(1,21):
  if i%3==0:
    print("Fizz")
  else:
    print(i)
```

Dla listy ["jabłko", "banan", "gruszka"] wypisz każdy element z numerem pozycji (indeks).

```python
lista = ["jabłko", "banan", "gruszka"]
for numer,owoc in enumerate(lista):
  print(numer, owoc)
```

```python
lista = ["jabłko", "banan", "gruszka"]
for i in range(len(lista)):
  print(i, lista[i])
```

5. Dla listy liczb wypisz tylko te, które są większe od średniej wszystkich liczb w liście.

```python
lista = input("Podaj liczby oddzielone spacją: ")
lista1=[int(x) for x in lista.split()]
if len(lista1) == 0:
    print("Nie ma liczb")
else:
    suma = sum(lista1)
    srednia = suma / len(lista1)

    print("Suma:", suma)
    print("Średnia:", srednia)

    print("Liczby większe od średniej:")
    for x in lista1:
        if x > srednia:
            print(x)

```

6. Narysuj prostokąt z gwiazdek * o wymiarach podanych przez użytkownika (np. 5x3).

```python
szerokosc = int(input("Podaj szerokość: "))
wysokosc = int(input("Podaj wysokość: "))

for i in range(wysokosc):
    print("*" * szerokosc)
```
