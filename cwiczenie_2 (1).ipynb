Poproś użytkownika o liczbę całkowitą i sprawdź, czy jest parzysta czy nieparzysta.

```python


def parzysta():
  p = int(input("Podaj liczby: "))
  if p % 2 == 0:
    print("Liczba jest parzysta")
  else:
    print("Liczba jest nieparzysta")
  print(p)

parzysta()
```

3.	Sprawdź, czy dana liczba jest liczbą pierwszą.


```python

def pierwsza():
  liczba = int(input("Podaj liczbe"))
  if liczba <= 1:
    print("liczba nie jest pierwsza")
  for i in range (2, liczba):
    if liczba % i == 0:
      print("Liczban nie jest pierwsza")
      break
    else:
      print("Liczba jest pierwsza")
      break




pierwsza()
```

4.	Zamień liczbę w systemie dziesiętnym na binarny, oktalny i heksadecymalny.

```python
n = 13

binary = bin(n)[2:]
octal = oct(n)[2:]
hexadecimal = hex(n)[2:].upper()

print(f"Decimal: {n}")
print(f"Binary: {binary}")
print(f"Octal: {octal}")
print(f"Hexadecimal: {hexadecimal}")
```


5.	Napisz program, który sprawdzi, czy liczba jest doskonała (suma jej dzielników = liczba).


```python
def doskonala():
  liczba = int(input("Podaj liczbe"))
  suma = 0
  for i in range (1, liczba-1):
    if liczba % i == 0:
      suma +=i
  if suma == liczba:
      print("Doskonala")
  else:
      print("niedoskonala")

doskonala()
```
