3.	Policz, ile razy dana liczba pojawia się w krotce.

```python
krotkax = (1,1,1,2,4,3,5,5,2)
def policz_w_krotce():
  suma=0
  x = int(input("Podaj licbe"))
  for i in krotkax:
    if i==x:
      suma+=1
  return suma


policz_w_krotce()
```

alternatywna wersja

```python
krotkax = (1,1,1,2,4,3,5,5,2)
x = int(input("Podaj liczbę: "))
print(krotkax.count(x))
```

5.	Zamień krotkę na listę, zmodyfikuj dane, a następnie z powrotem na krotkę.

```python
krotka = (1, 2, 3, 4, 5)
print("Oryginalna krotka:", krotka)
lista = list(krotka)
print("Po zamianie na listę:", lista)
# zmiana elementu
lista[0] = 10

# dodanie elementu
lista.append(6)

# usunięcie elementu
lista.remove(3)

print("Zmodyfikowana lista:", lista)
krotka = tuple(lista)
print("Nowa krotka:", krotka)

```
