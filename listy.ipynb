4.	Oblicz sumę wszystkich liczb w liście.

```python
lista = [random.randint(1, 200) for _ in range(10)]
suma = sum(lista)
print(lista)
print("Suma elementów:", suma)
```

5.	Posortuj listę rosnąco i malejąco.


```python
lista = [random.randint(1, 200) for _ in range(10)]
print(lista)
lista_rosnaco = sorted(lista)
print("rosnaco", lista_rosnaco)
lista_malejaco = sorted(lista, reverse=True)
print("malejaco", lista_malejaco)
```
