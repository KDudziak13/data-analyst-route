```python
1.	Poproś użytkownika o dwie liczby zmiennoprzecinkowe i oblicz ich średnią.
```

```python
def srednia():
  liczba1 = float(input("Podaj pierwsza liczbe"))
  liczba2 = float(input("Podaj pdruga liczbe"))
  s= (liczba1+liczba2)/2
  print("srednia", s)
srednia()
```

2.	Oblicz pierwiastek kwadratowy liczby zmiennoprzecinkowej bez użycia math.sqrt().

```python
def pierwiastek():
  liczba = float(input("Podaj liczbe"))
  pierwiastek = liczba**0.5
  print("pierwiastek", round(pierwiastek, 2))

pierwiastek()
```


3.	Zamień temperaturę z Celsjusza na Fahrenheita i odwrotnie.

```python
def cels():
  c = float(input("Podaj temp w C"))
  f = (c*9/5+32)
  print("temp w F", f)

cels()
```

```python
def far():
  f = float(input("Podaj temp w F"))
  c = ((5/9) * (f-32))
  print("temp w C", c)

far()
```

4.	Poproś użytkownika o liczbę i sprawdź, czy jest dodatnia, ujemna czy zerowa.

```python
def czy():
  liczba = float(input("Podaj liczbe"))
  if liczba > 0:
    print("Liczba jest dodatnia")
  elif liczba < 0:
    print("Liczba jest ujemna")
  else:
    print("Liczba jest zerowa")

czy()
```

```python
def f2():
  x = float(input("Podaj liczbe x"))
  a=2.4
  b=3.2
  c=-4.5

  fx = a*x**2 + b*x + c
  print("wartosc f(x)", round(fx, 2))


f2()
```
