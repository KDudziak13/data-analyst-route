```python
5.	Policz ile razy każde słowo występuje w podanym zdaniu i zapisz wynik w słowniku.
```

```python
zdanie = input("Podaj zdanie")
lista = zdanie.split()
wynik = {}

for slowa in lista:
  if slowa in wynik:
    wynik[slowa] +=1
  else:
    wynik[slowa] =1
print(wynik)
```
