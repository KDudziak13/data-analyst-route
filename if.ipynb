1.	Sprawdź, czy liczba podana przez użytkownika jest dodatnia, ujemna czy zero.

```python
liczba = int(input("Podaj liczbe: "))
if liczba == 0:
  print("Twoja liczba to 0")
elif liczba > 0:
  print("Twoja liczba to dodatnia")
else:
  print("Twoja liczba to ujemna")
```

Sprawdź, czy liczba jest parzysta czy nieparzysta.

```python
liczba = int(input("Podaj liczbe: "))
if liczba % 2 == 0:
  print("Twoja liczba jest parzysta")
else:
  print("Twoja liczba jest nieparzysta")
```

Sprawdź, czy użytkownik jest pełnoletni (wiek >= 18) i wyświetl odpowiedni komunikat.

```python
wiek = int(input("Podaj swoj wiek: "))
if wiek <= 0:
  print("Bledny wiek")
elif wiek >= 18:
  print("Jestes pelnoletni")
else:
  print("Jestes niepelnoletni")
```

Dla liczby od 1 do 100, jeśli jest podzielna przez 3 wypisz "Fizz", jeśli przez 5 "Buzz", jeśli przez oba "FizzBuzz".

```python

for i in range(1,100):
  if i % 3 == 0 and i % 5 == 0:
    print("FizzBuzz")
  elif i % 3 == 0:
    print("Fizz")
  elif i % 5 == 0:
    print("Buzz")
  else:
    print(i)
```

Napisz funkcję, która przyjmuje wynik testu (0–100) i zwraca ocenę literową (A, B, C, D, F).

```python
wynik = int(input("Podaj wynik testu: "))
if wynik < 0 or wynik > 100:
  print("Bledny wynik")
elif wynik <30:
  print("Ocena F")
elif wynik <45:
  print("Ocena D")
elif wynik <60:
  print("Ocena C")
elif wynik <80:
  print("Ocena F")
else:
  print("Ocena F")
```

Napisz program, który ocenia temperaturę <0 → Mróz, 0–10 → "Chłodno", 11–20 → "Umiarkowanie", >20 → "Ciepło".

```python
temperatura = int(input("Podaj temperature: "))
if temperatura < 0:
  print("Mróz")
elif temperatura < 10:
  print("Chlodno")
elif temperatura < 20:
  print("Umiarkowanie")
else:
  print("Cieplo")
```
